```python
#Import Packages
from selenium import webdriver
import time
import pandas as pd
```

```python
#Import Packages

from selenium.webdriver.support.select import Select
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service as ChromeService
```

```python
def web_driver():
    options = webdriver.ChromeOptions()
    options.add_argument("--verbose")
    options.add_argument('--no-sandbox')
    options.add_argument('--headless')
    options.add_argument('--disable-gpu')
    options.add_argument("--window-size=1920, 1200")
    options.add_argument('--disable-dev-shm-usage')
    driver = webdriver.Chrome(options=options)
    return driver
```

```python
url = 'https://www.linkedin.com/jobs/search?keywords=ML%20Engineer&location=India&locationId=&geoId=102713980&f_TPR=&f_PP=106888327&f_E=2&position=1&pageNum=0'
```

```python
# Automated Chrome Browser Doesn't work in colab
driver = web_driver()
driver.get(url)
driver.page_source
```

```python
# Total job count From the url
job_count=driver.find_elements(By.CLASS_NAME,'results-context-header__job-count')[0].text
```

```python
job_count =  int(job_count.replace(",", "").replace("+", ""))
print(job_count)
```

```python
i = 2
# Continue scrolling until a certain condition is met
while i <= int((job_count + 200) / 25) + 1:
    # Scroll to the bottom of the page
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    i = i + 1

    try:
        # Attempt to find and click the "Load more results" button
        send = driver.find_element_by_xpath("//button[@aria-label='Load more results']")
        driver.execute_script("arguments[0].click();", send)
        time.sleep(3)

    except:
        pass

    # Add a delay to avoid making too many requests in a short time
    time.sleep(5)
```

```python
# Storing the INFO
companyname=[]
titlename=[]
locations=[]
href_list=[]
recurt_info=[]
```

```python
# Extracting company names
try:
    for i in range(job_count):
        company=driver.find_elements(By.CLASS_NAME,'base-search-card__subtitle')[i].text
        companyname.append(company)

except IndexError:
    print()
```

```python
# Extracting Job Ttile
try:
    for i in range(job_count):
        title=driver.find_elements(By.CLASS_NAME,'base-search-card__title')[i].text
        titlename.append(title)

except IndexError:
    print()
```

```python
#Extracting Location

try:
    for i in range(job_count):
        location=driver.find_elements(By.CLASS_NAME,'job-search-card__location')[i].text
        locations.append(location)


except IndexError:
    print()
```

```python
#Extracting Job Links
try:
    jobList = driver.find_elements(By.CLASS_NAME,'base-card__full-link')
    href_list=[]
    for e in jobList:
        href_list.append(e.get_attribute('href'))

except IndexError:
    print()
href_list.append("ggggg")
href_list.append("ggghhh")
href_list.append("ggghhsh")
```

```python
print(len(companyname), len(titlename), len(locations), len(href_list))
```

```python
data = {
    "companyname": companyname,
    "titlename": titlename,
    "locations": locations,
    "href_list": href_list

}

# Create the DataFrame
df = pd.DataFrame(data)

print(df.head())
```
